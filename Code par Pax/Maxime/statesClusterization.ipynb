```python
df_submission = pd.read_csv("Data/03_Prediction/States_Pred.csv")
states = df_submission["State"].unique()
```

```python
df_Rabi = pd.read_csv("Data/Results/GP_Pred_Rabi_23-02.csv")
```

```python
results_Rabi = {}

for state in states:
    # get the values count of clusters by state
    df_state = df_Rabi[df_Rabi['State'] == state]
    clusters = df_state['Cluster'].value_counts(ascending=False)
    # print(clusters.to_dict())
    # print(clusters)
    results_Rabi[state] = clusters.to_dict()

results_Rabi
```

```python
df_Kharif = pd.read_csv("Data/Results/GP_Pred_Kharif_23-02.csv")
```

```python
results_Kharif = {}

for state in states:
    # get the values count of clusters by state
    df_state = df_Kharif[df_Kharif['State'] == state]
    clusters = df_state['Cluster'].value_counts(ascending=False)
    # print(clusters.to_dict())
    # print(clusters)
    results_Kharif[state] = clusters.to_dict()

results_Kharif
```

```python
from Environnement.extractClusters import get_cluster_states

import numpy as np

def get_cluster_states(df, state, rule="max"):
    
        """
        Return the cluster of a given key using the query dataframe based on a specific rule.
        """
        clusters_values = df[df["State"]==state]["Cluster"].value_counts(ascending=False, normalize=True)
    
        if rule == "max":
            # We return the cluster with the highest number of values
            # print(clusters_values.values.argmax())
            # print(clusters_values.index)
            # print(clusters_values.index[clusters_values.values.argmax()])
            
            # if the type of clusters_values.values.argmax() is int, we return the index
            print(state)
            print(clusters_values)
            print(isinstance(clusters_values.values.argmax(), np.int64))

            if isinstance(clusters_values.values.argmax(), np.int64):
                print('oui')
                return clusters_values.index[clusters_values.values.argmax()]

            else:
                size = len(clusters_values.values.argmax())

                return clusters_values.index[clusters_values.values.argmax()[np.random.randint(0, size)]]
        
        elif rule == "draw":
            return np.random.choice(clusters_values.index, 1, p=clusters_values.values)
    
        else:
            assert False, "Unknown rule"


data = {'Kharif': df_Kharif, 'Rabi': df_Rabi}

for season in data.keys():
    clusters = []
    df = data[season]
    for state in states:
        clusters.append(get_cluster_states(df, state, rule="max"))

    df_submission[season] = clusters
    

df_submission
```
